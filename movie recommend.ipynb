data preprocessing

```python
movies['year']=movies['title'].str.extract('(\(\d\d\d\d\))',expand=False)
```

```python
movies['year']=movies['year'].str.extract('(\d\d\d\d)',expand=False)
```

```python
movies.head()
```

```python
movies['title']=movies['title'].str.replace('(\(\d\d\d\d\))' , '')
```

```python
movies['title']=movies['title'].apply(lambda x: x.strip())
```

```python
movies['genres']=movies['genres'].str.split('|')
```

```python
movies.head()
```

```python
movieswithgenres=movies.copy()
```

```python
for index,row in movies.iterrows():
    for genres in row['genres']:
        movieswithgenres.at[index,genres]=1
movieswithgenres=movieswithgenres.fillna(0)       
movieswithgenres.head()        
```

```python
userinput=[{'title':'Jumanji' , 'rating':'3.5'},
           {'title':'Toy Story' , 'rating':'4'},
           {'title':'Grumpier Old Men', 'rating':'3'}
          ]
inputmovie=pd.DataFrame(userinput)
inputmovie
```

```python
inputid=movies[movies['title'].isin(inputmovie['title'])]
```

```python
inputmovie=pd.merge(inputid,inputmovie)
```

```python
inputmovie
```

```python
inputmovie.drop(['genres','year'], 1)
```

now we are ready with user's input and we have to fetch their preferred genres

```python
usermovie=movieswithgenres[movieswithgenres['movieId'].isin(inputmovie['movieId'])]
```

```python
usermovie
```

```python
usermovie=usermovie.reset_index(drop=True)
usermovie
```

```python
 usergenres=usermovie.drop(['movieId','title','genres','year'], 1)
```

```python
usergenres
```

```python
inputmovie['rating']=inputmovie['rating'].astype(float)
inputmovie['rating']
```

now we are ready to learn user's preference 

```python
userprofile=usergenres.transpose().dot(inputmovie['rating'])
```

```python
userprofile
```

now this is the profile of the user which we will use to recommend movies to the user

```python
genretable=movieswithgenres.drop(['title','year','genres'], 1)

genretable=genretable.set_index(['movieId'])
genretable
```

```python
recommendationtable=((genretable*userprofile).sum(axis=1)/userprofile.sum())
```

```python
recommendationtable.head()
```

this is the the final recommendation table

```python
movies.loc[movies['movieId'].isin(recommendationtable.head(20).keys())]
```
